```python
import pandas as pd

job_lists = [
        'job_lists/arm_server_1.5GHz_500', 
        'job_lists/arm_server_1.5GHz_502',
        'job_lists/arm_server_1.5GHz_505',
        'job_lists/arm_server_1.5GHz_520',
        'job_lists/arm_server_1.5GHz_523',
        'job_lists/arm_server_1.5GHz_525',
        'job_lists/arm_server_1.5GHz_531',
        'job_lists/arm_server_1.5GHz_541',
        'job_lists/arm_server_1.5GHz_548',
        'job_lists/arm_server_1.5GHz_557',
        'job_lists/arm_server_1.5GHz_503',
        'job_lists/arm_server_1.5GHz_507',
        'job_lists/arm_server_1.5GHz_508',
        'job_lists/arm_server_1.5GHz_510',
        'job_lists/arm_server_1.5GHz_511',
        'job_lists/arm_server_1.5GHz_519',
        'job_lists/arm_server_1.5GHz_521',
        'job_lists/arm_server_1.5GHz_527',
        'job_lists/arm_server_1.5GHz_538',
        'job_lists/arm_server_1.5GHz_544',
        'job_lists/arm_server_1.5GHz_549',
        'job_lists/arm_server_1.5GHz_554',
        'job_lists/arm_server_3.0GHz_500', 
        'job_lists/arm_server_3.0GHz_502',
        'job_lists/arm_server_3.0GHz_505',
        'job_lists/arm_server_3.0GHz_520',
        'job_lists/arm_server_3.0GHz_523',
        'job_lists/arm_server_3.0GHz_525',
        'job_lists/arm_server_3.0GHz_531',
        'job_lists/arm_server_3.0GHz_541',
        'job_lists/arm_server_3.0GHz_548',
        'job_lists/arm_server_3.0GHz_557',
        'job_lists/arm_server_3.0GHz_503',
        'job_lists/arm_server_3.0GHz_507',
        'job_lists/arm_server_3.0GHz_508',
        'job_lists/arm_server_3.0GHz_510',
        'job_lists/arm_server_3.0GHz_511',
        'job_lists/arm_server_3.0GHz_519',
        'job_lists/arm_server_3.0GHz_521',
        'job_lists/arm_server_3.0GHz_527',
        'job_lists/arm_server_3.0GHz_538',
        'job_lists/arm_server_3.0GHz_544',
        'job_lists/arm_server_3.0GHz_549',
        'job_lists/arm_server_3.0GHz_554'
]

num_counters = 4

for job_list in job_lists:
    input_file = open(job_list, 'r')
    csv_files = input_file.readlines()

    merged_df = pd.read_csv('blank.csv')
    initial_read = 0
    
    for csv in csv_files:
        csv = csv.strip()
        df1 = pd.read_csv(csv)
        df1['time'] = df1['time'].astype(float).round(2)
        df1.drop(df1.columns[4:], axis=1, inplace=True)
        df1.drop(df1.columns[2], axis=1, inplace=True)

        # Get counter 1
        df2 = df1.iloc[num_counters-0::4, :]
        counter_name = df2.iloc[num_counters-0][2]
        df2 = df2.rename(columns={'value': counter_name})
        df2.drop(df2.columns[2], axis=1, inplace = True)
        df2 = df2.round({'time':2})
       
        # Get Counter 2
        df3 = df1.iloc[num_counters-3::4, :]
        counter_name = df3.iloc[num_counters-0][2]
        df3 = df3.rename(columns={'value': counter_name})
        df3.drop(df3.columns[2], axis=1, inplace = True)
        df3 = df3.round({'time':2})
        
        # Get Counter 3
        df4 = df1.iloc[num_counters-2::4, :]
        counter_name = df4.iloc[num_counters-0][2]
        df4 = df4.rename(columns={'value': counter_name})
        df4.drop(df4.columns[2], axis=1, inplace = True)
        df4 = df4.round({'time':2})
        
        # Get Counter 4
        df5 = df1.iloc[num_counters-1::4, :]
        counter_name = df5.iloc[num_counters-0][2]
        df5 = df5.rename(columns={'value': counter_name})
        df5.drop(df5.columns[2], axis=1, inplace = True)
        df5 = df5.round({'time':2})

        # Merge DFs
        df6 = pd.merge(df2, df3, on = 'time', how='inner')
        df6 = df6.round({'time':2})
        df7 = pd.merge(df4, df5, on='time', how='inner')
        df7 = df7.round({'time':2})
        df8 = pd.merge(df6, df7, on='time', how='inner')
        df8 = df8.round({'time':2})

        if initial_read == 0:
            merged_df = pd.merge(merged_df, df8, on='time', how = 'inner')
            initial_read = 1
        else:
            df8 = df8.drop(columns=['instructions'])
            merged_df = pd.merge(merged_df, df8, on='time',how='inner')
            
    merged_df.to_csv('traces/aligned_' + job_list[10:], sep=',', index=False)
```
