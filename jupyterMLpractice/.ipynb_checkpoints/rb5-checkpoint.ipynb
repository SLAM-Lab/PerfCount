```python
from matplotlib import pyplot as plt 
from sklearn.pipeline import make_pipeline
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,  GridSearchCV
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import PolynomialFeatures
```

```python
inOrder = pd.read_csv("arm_edge_in_order_cpu.csv")
ooBig = pd.read_csv("arm_edge_out_of_order_cpu_large_cache.csv")
oo = pd.read_csv("arm_edge_out_of_order_cpu.csv")

#remove outliers
```

```python
col_names = ['Workload']
res_name = ['IPC']
xInOrder = inOrder.loc[:,col_names].values
yInOrder = inOrder.loc[:,res_name].values

xOObig = ooBig.loc[:,col_names].values
yOObig = ooBig.loc[:,res_name].values

xOO = oo.loc[:,col_names].values
yOO = oo.loc[:,res_name].values
```

```python
plt.scatter(xInOrder, yInOrder, color='blue', label='In Order')
plt.title('In Order')
plt.xlabel('Workload')
plt.ylabel('IPC')
plt.grid()
```

```python
plt.scatter(xOObig, yOObig, color='blue', label='Out of Order Large')
plt.title('Out of Order Large')
plt.xlabel('Workload')
plt.ylabel('IPC')
plt.grid()
```

```python
plt.scatter(xOO, yOO, color='blue', label='Out of Order')
plt.title('Out of Order')
plt.xlabel('Workload')
plt.ylabel('IPC')
plt.grid()
```

```python
#RNN (LSTM), MLP, SVM, Perceptrons 

bigsx = yOObig
bigsy = yOO

# Split the data
X_train, X_test, y_train, y_test = train_test_split(bigsx, bigsy, test_size=0.2, random_state=42)

#test both the big cores

# Train the model
model = LinearRegression()
model.fit(X_train.reshape(-1, 1), y_train)
y_pred = model.predict(X_test.reshape(-1, 1))
print("R-squared:", r2_score(y_test, y_pred))
```

```python
# x = yInOrder
# y = yOO.ravel()
from sklearn.ensemble import GradientBoostingRegressor,VotingRegressor
from xgboost import XGBRegressor

X = yInOrder
y = yOO.ravel()

# rf_model = RandomForestRegressor(n_estimators=100)
# rf_model.fit(X, y)
# # Predict
# rf_predictions = rf_model.predict(X)
# # Calculate R²
# rf_r2 = r2_score(y, rf_predictions)
# print(f"Random Forest R² value: {rf_r2}")
# plt.scatter(X, y, color='blue', label='Actual')
# plt.scatter(X, rf_predictions, color='red', label='Predicted')
# plt.xlabel('X values')
# plt.ylabel('Y values')
# plt.title('Random Forest Predicted vs Actual Values')
# plt.legend()
###############################################

#ensemble approach 
# rf_model = RandomForestRegressor(n_estimators=100)
# gb_model = GradientBoostingRegressor(n_estimators=100)

# # Combine models into a voting regressor
# ensemble_model = VotingRegressor([('rf', rf_model), ('gb', gb_model)])
# # Fit the ensemble model
# ensemble_model.fit(X, y)
# # Predict
# ensemble_predictions = ensemble_model.predict(X)
# # Calculate R²
# ensemble_r2 = r2_score(y, ensemble_predictions)
# print(f"Ensemble R² value: {ensemble_r2}")
# # Plot actual vs predicted values
# plt.scatter(X, y, color='blue', label='Actual')
# plt.scatter(X, ensemble_predictions, color='red', label='Predicted')
# plt.xlabel('X values')
# plt.ylabel('Y values')
# plt.title('Ensemble Predicted vs Actual Values')
# plt.legend()
##################################################

# Create individual models
rf_model = RandomForestRegressor(n_estimators=200)
gb_model = GradientBoostingRegressor(n_estimators=200)
xgb_model = XGBRegressor(n_estimators=200)
# Combine models into a voting regressor
ensemble_model = VotingRegressor([('rf', rf_model), ('gb', gb_model), ('xgb', xgb_model)])
# Fit the ensemble model
ensemble_model.fit(X, y)
# Predict
ensemble_predictions = ensemble_model.predict(X)
# Calculate R²
ensemble_r2 = r2_score(y, ensemble_predictions)
print(f"Advanced Ensemble R² value: {ensemble_r2}")
# Plot actual vs predicted values
plt.scatter(X, y, color='blue', label='Actual')
plt.scatter(X, ensemble_predictions, color='red', label='Predicted')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Advanced Ensemble Predicted vs Actual Values')
```

```python
x_show = np.array(X).flatten()
y_show = np.array(y).flatten()
ensemble_predictions_show = np.array(ensemble_predictions).flatten()

data = {'X values': x_show, 'Actual values': y_show, 'Predicted values': ensemble_predictions_show}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
